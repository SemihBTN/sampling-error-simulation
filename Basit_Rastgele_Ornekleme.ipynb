```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Ortalaması 50 , sapması 15 olan 10000 kişilik bir popülasyon oluşturuyoruz.
population = np.random.normal(50, 15, 10000)
gercek_ort = population.mean() # Popülasyon ortalamasını buluyoruz.
boyutlar = [10, 50, 100, 500, 1000, 5000]
hatalar = [] # Ana listemiz

for n in boyutlar:
    gecici_hatalar = [] # Her boyut için yeni bir liste
    for i in range(50): # 50 deneme yapıyoruz
        orneklem = np.random.choice(population, size=n, replace=False)
        orneklem_ort = orneklem.mean()
        hata = abs(gercek_ort - orneklem_ort)
        gecici_hatalar.append(hata)
    
    # DİKKAT: Burası for i döngüsünün dışı, for n döngüsünün içi!
    # 50 denemenin ortalamasını alıp ana listeye ekliyoruz
    hatalar.append(np.mean(gecici_hatalar)) 

# Çizerken 'hatalar' listesini kullanıyoruz
plt.plot(boyutlar, hatalar, marker="o", color="r")
plt.title("Boyutun Hataya Olan Etkisi (Pürüzsüz)")
plt.xlabel("Boyutlar")
plt.ylabel("Hata Oranı")
plt.show()
```
